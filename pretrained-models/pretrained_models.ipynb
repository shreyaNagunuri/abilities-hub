# EyePop Person Detection Demo Notebook

## Overview

This notebook demonstrates **real-time person detection** using the EyePop SDK.  
It is designed for developers who need instant, reliable person localization in images, video, or live streams—without any ML overhead.

---

## What It Does

- Detects people in images, recorded video, or live streams
- Returns structured bounding box coordinates with confidence scores
- No training, no configuration—just provide media and your API key

---

## Why Use This Model?

Most person detection workflows fail because they:
- Require custom training and dataset prep
- Are too slow for real-time applications

EyePop removes both constraints, giving you reliable person localization immediately.

Bounding boxes are the atomic unit for:
- Counting
- Tracking
- Behavior analysis
- Access control
- Safety monitoring

```python
import getpass

EYEPOP_API_KEY=getpass.getpass('Enter your API KEY: ')
```

```python
from pathlib import Path
from eyepop import EyePopSdk
from eyepop.worker.worker_types import InferenceComponent, Pop
import json


pop = Pop(components=[
   InferenceComponent(
       ability=f"eyepop.person:latest"
   )
])

with EyePopSdk.workerEndpoint(api_key=EYEPOP_API_KEY) as endpoint:
   endpoint.set_pop(pop)
   sample_img_path = Path("test.png")
   job = endpoint.upload(sample_img_path)
   while result := job.predict():
      print(json.dumps(result, indent=2))

print("Done")
```
